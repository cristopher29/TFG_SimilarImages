### Data processing

```python
import pandas as pd
```

```python
xl = pd.ExcelFile('data.xlsx')
type(xl)
```

```python
df = xl.parse('Full1')
df.head() 
```

```python
df['name'] = df[100].combine_first(df[110])
df.drop(110, axis=1, inplace=True)
df.drop(100, axis=1, inplace=True)
df.head() 
```

```python
df = df.rename(columns={'859|c': 'tag', '859|u': 'filepath'})
df.head()
```

```python
import unicodedata

def strip_accents(text):
    try:
        text = unicode(text, 'utf-8')
    except NameError:
        pass
    text = unicodedata.normalize('NFD', text)\
           .encode('ascii', 'ignore')\
           .decode("utf-8")
    return str(text)

def f1(x):
    return list(set(list(filter(None, strip_accents(x).lower().replace(',',' ').replace('"',' ').replace(' de ',' ').replace(' del ',' ').split(' ')))))

df['tag'] = df['tag'].apply(lambda x: f1(x) if pd.notnull(x) else x)
```

```python
df.head()
```

```python
df['filepath'] = df['filepath'].apply(lambda x: x.replace('"','').split(','))
df.head()
```

```python
df = df[df['tag'].notna()]
df = df.reset_index(drop=True)
df.head()
```

```python
def f_rename(urls):
    new = list()
    for url in urls:
        new.append(url[url.rfind("/")+1:])
    return new

df['filepath'] = df['filepath'].apply(lambda urls: f_rename(urls))
df.head()
```

```python
df_split = pd.DataFrame(columns=['tag', 'filepath', 'autor'])
count = 0
for index, row in df.iterrows():
    for url in row['filepath']:
        df_split.loc[count] = [row['tag'], url, row['name']]
        count+=1
df = df_split
df.head()
```

```python
import numpy as np
import itertools
from collections import Counter
from matplotlib import pyplot as plt

tags = df['tag'].to_numpy()
tags = list(itertools.chain.from_iterable(tags))

tags_counts = Counter(tags)
df_tags = pd.DataFrame.from_dict(tags_counts, orient='index')
df_tags.columns = ['count']
df_tags = df_tags.sort_values(by='count', ascending=False)
print(len(df_tags))
df_tags.head(50)
```

```python
df.to_csv('pre_process.csv', index=False, encoding='utf-8')
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

df['tag'] = df['tag'].apply(lambda tags: tuple(tags))

mlb = MultiLabelBinarizer()
mlb.fit(df['tag'])
df["tag_bina"] = [[] for _ in range(len(df))]
df['tag_bina'] = mlb.transform(df['tag']).tolist()
print(len(df['tag_bina'][0]))
df.head()
```

```python
df = df.drop_duplicates(subset='filepath', keep="last")
df = df.reset_index(drop=True)

df.to_pickle('data_processing.pkl')
```
